Português: Utilização de histogramas de colunas para analisar distorção das variáveis de característica dos vinhos (Acidez fixa, total de dióxido de enxofre, pH, álcool); e gráficos de dispersão de qualidade para analisar o impacto das variáveis acidez volátil, açúcar residual, pH e álcool na qualidade dos vinhos.

English: Use of column histograms to analyze distortion of wine characteristic variables (fixed acidity, total sulfur dioxide, pH, alcohol); and quality scatter plots to analyze the impact of volatile acidity, residual sugar, pH and alcohol variables on wine quality.

```python
import pandas as pd
import numpy as np

%matplotlib inline

df = pd.read_csv('winequality_edited.ipynb.csv')
display(df.head())
```

```python
df.hist(figsize=(10,10))
```

A acidez é uma das características mais atraentes de um vinho. 
Ela é responsável pela sensação de frescor da bebida, além de mostrar a exuberância das frutas e a complexidade de sabores presentes no vinho.

Acidity is one of the most attractive characteristics of a wine. It is responsible for the sensation of freshness of the drink, in addition to showing the exuberance of the fruits and the complexity of flavors present in the wine.

```python
df["fixed acidity"].plot(kind="hist", title="Acidez");
```

A análise do histograma de acidez mostra uma distorção à direita, onde um grande número de amostras apresenta acidez entre os valores 6 e 8.

The analysis of the acidity histogram shows a distortion to the right, where a large number of samples have acidity between values 6 and 8.

No vinho, o dióxido de enxofre (SO2) é utilizado como conservante, devido às suas propriedades antioxidantes e antibacterianas, evitando que os vinhos degradem de forma mais rápida.

In wine, sulfur dioxide (SO2) is used as a preservative, due to its antioxidant and antibacterial properties, preventing wines from degrading more quickly.

```python
df["total sulfur dioxide"].plot(kind="hist", title="Dióxido de Enxofre"); 
```

A análise do histograma de níveis de dióxido de enxofre mostra um grande número de amostras apresentando níveis de dióxido de enxofre entre 100 e 200.

Analysis of the histogram of sulfur dioxide levels shows a large number of samples showing sulfur dioxide levels between 100 and 200.

O pH no vinho afeta o sabor do produto final, sua cor, oxidação e estabilidade química. Eis porque medi-lo, e intervir quando necessário, se torna fundamental.
Essa medida fornece uma indicação do grau de maturação das uvas e composição microbiana e química do mosto e do próprio vinho. Ao realizar-se a medição do PH, pode-se impedir o desenvolvimento de microrganismos superficiais indesejados.

The pH of wine affects the flavor of the final product, its color, oxidation and chemical stability. This is why measuring it and intervening when necessary is essential.
This measurement provides an indication of the degree of ripeness of the grapes and the microbial and chemical composition of the must and the wine itself. By measuring the pH, the development of unwanted surface microorganisms can be prevented.

```python
df["pH"].plot(kind="hist", title="pH");
```

A análise dos valoresse pH visualizados no histograma mostra uma maior frequência de amostras com pH em torno de 3.2.

Analysis of the pH values displayed in the histogram shows a higher frequency of samples with a pH around 3.2.

O álcool existente nos vinhos e espumantes é o elemento que fornecerá a capacidade de conservação, personalidade e longevidade aos vinhos. 

The alcohol present in wines and sparkling wines is the element that provides the wines with their conservation capacity, personality and longevity.

```python
df["alcohol"].plot(kind="hist", title="Alcohol");
```

Percebe-se que os dados de teor alcoólico apresentam no histograma um deslocamento à direita, indicando uma frequência maior em amostras de teor alcólico com graduação entre 9 e 11.

The alcohol content histogram shows a shift to the right, indicating a higher frequency in samples with alcohol content between 9 and 11.

```python
df.plot(x='volatile acidity', y='quality', kind='scatter');
```

```python
df.plot(x='residual sugar', y='quality', kind='scatter');
```

```python
df.plot(x='pH', y='quality', kind='scatter');
```

```python
df.plot(x='alcohol', y='quality', kind='scatter');
```

Apesar de não ficar totalmente claro, conforme o valor de alcohol aumeta, temos uma tendência a aumentar a qualidade. Podemos ver como o deslocamento das linhas da qualidade tendem para a direita.

Pode-se verificar as correlações através do método de Pearson:

Although it is not entirely clear, as the alcohol value increases, we have a tendency to increase the quality. We can see how the quality lines move to the right.

The correlations can be verified using the Pearson´s method:

```python
data = df[['volatile acidity','quality']]
data.corr(method='pearson')
```

```python
data2 = df[['residual sugar','quality']]
data2.corr(method='pearson')
```

```python
data3 = df[['pH','quality']]
data3.corr(method='pearson')
```

```python
data3 = df[['alcohol','quality']]
data3.corr(method='pearson')
```

O coeficiente de correlação de Pearson mede a relação estatística entre duas variáveis contínuas, e varia em um intervalo entre -1 e +1. 
valor 0: não há associação entre as duas variáveis; 
valor > 0 indica uma associação positiva (à medida que o valor de uma variável aumenta, o mesmo acontece com o valor da outra variável);
valor < 0 indica uma associação negativa (à medida que o valor de uma variável aumenta, o valor da outra diminui).


Pearson's correlation coefficient measures the statistical relationship between two continuous variables, and varies in a range between -1 and +1.
value 0: there is no association between the two variables;
value > 0 indicates a positive association (as the value of one variable increases, so does the value of the other variable);
value < 0 indicates a negative association (as the value of one variable increases, the value of the other decreases).
