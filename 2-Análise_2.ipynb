É possível combinar os conjuntos de dados sobre vinho tinto e branco para que a análise seja mais eficiente.
NumPy foi usado para criar uma nova coluna que preserva as informações de cores e, então, o Pandas para combinar os dataframes.

It is possible to combine the red and white wine datasets for more efficient analysis.
NumPy was used to create a new column that preserves the color information, and then Pandas was used to combine the dataframes.

```python
import pandas as pd
import numpy as np

df_red = pd.read_csv('winequality-red.csv', sep=';')
display(df_red.head(5))

df_white = pd.read_csv('winequality-white.csv', sep=';')
display(df_white.head(5))
```

Funcão loc[linhas,colunas].
O operador [:] representa o total da lista (ou string).
Ou seja, para todas as linhas da tabela df_red.loc, cria-se uma coluna nomeada 'color', e preenche-se as linhas dessa coluna com 'red'.

Function loc[rows,columns].
The operator [:] represents the total of the list (or string).
For all the rows in the df_red.loc data set, a column named 'color' is created, and the rows in that column are filled with 'red'.

```python
df_red.loc[:,'color']=('red')
display(df_red.head(5))
```

```python
df_white.loc[:,'color']=('white')
display(df_white.head(5))
```

Outra forma de adicionar a coluna 'color' ao dataframe (Another way off adding the 'color' column to the dataframe):

```python
# criando vetor de cor para o dataframe tinto (creating color vector for red dataframe)
color_red = np.repeat("red", df_red.shape[0])
print(color_red)

# criando vetor de cor para o dataframe branco (creating color vector for white dataframe)
color_white = np.repeat("white", df_white.shape[0])
print(color_white)

#np.repeat("red", red_df.shape[0]) função de comando para o numpy repetir 'red' o número de linhas do dataframe (rows)
#np.repeat ("red", red_df.shape[0]) command function for numpy to repeat 'red' the number of rows of the dataframe
#shape() consists of array having two arguments rows and columns
#if you search shape[0] then it will gave you the number of rows (linhas)
```

```python
df_red['color'] = color_red
df_red.head()
```

```python
df_white['color'] = color_white
df_white.head()
```

```python
# anexando dataframes
wine_df = pd.concat([df_red, df_white])

# exibindo o data frame para ver se tudo deu certo (displaying data frame)
#df.loc[row,column]
#display(wine_df.loc[])
#shape mostra as dimensões (linhas e colunas) do data frame
print(wine_df.shape)
```
